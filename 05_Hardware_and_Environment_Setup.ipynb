# 05. Hardware & Environment Setup

Fine-tuning is **maths on a lot of numbers**, so it runs best on a **GPU** (a graphics card).
Let's understand the hardware and install the libraries you need.

## Real-life analogy

Training a model on a **CPU** is like digging a pool with a **spoon**.
Training on a **GPU** is like using an **excavator**. Same job - a GPU is just built for it.

> No GPU on your laptop? Use **Google Colab** (free GPU) or **Kaggle Notebooks**. Easiest way to start.

## What you need

| Thing | What it is | Why it matters |
|-------|-----------|----------------|
| **GPU** | Graphics card (e.g. NVIDIA T4, A100) | Does the training fast |
| **CUDA** | NVIDIA's software to use the GPU | Lets PyTorch talk to the GPU |
| **VRAM** | Memory on the GPU | Bigger models need more VRAM |
| **QLoRA** | 4-bit training trick | Fits big models in **small** VRAM |

### Libraries we use

| Library | Job |
|---------|-----|
| `transformers` | Load models and tokenizers |
| `datasets` | Load and prepare data |
| `peft` | LoRA / QLoRA (parameter-efficient tuning) |
| `trl` | `SFTTrainer` - easy training |
| `bitsandbytes` | 4-bit quantization (for QLoRA, needs GPU) |
| `accelerate` | Runs training on the GPU smoothly |

```python
# Install everything (run once). On Colab, put a "!" in front of pip.
# !pip install transformers datasets peft trl bitsandbytes accelerate

# Now check what hardware you have:
import torch

print("PyTorch version:", torch.__version__)
print("GPU available? ", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU name:      ", torch.cuda.get_device_name(0))
else:
    print("No GPU found. Small demos still work on CPU (slowly).")
    print("Tip: use Google Colab and pick Runtime -> Change runtime type -> GPU.")
```

## Key points to remember

- Fine-tuning runs best on a **GPU**; a **CPU** works only for tiny demos (slow).
- **CUDA** lets PyTorch use an NVIDIA GPU; **VRAM** is the GPU's memory.
- **QLoRA** (4-bit) lets big models fit in **small VRAM**.
- No GPU? Use **Google Colab** (free) - set Runtime to **GPU**.
- Check your setup with `torch.cuda.is_available()`.
