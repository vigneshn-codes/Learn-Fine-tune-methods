# 14. Merging LoRA Weights & Saving the Model

After training, you have two pieces: the **base model** + the small **LoRA adapter**.
**Merging** combines them into **one normal model** that is easy to share and run.

## Real-life analogy

The LoRA adapter is like **sticky notes** on a book. Merging is like **printing a new book**
with those notes already included - now it is a single, complete book.

## Why merge?

| Before merging | After merging |
|----------------|---------------|
| Need base model **+** adapter | **One** self-contained model |
| Two things to load | Load **one** folder |
| Good while experimenting | Good for **sharing / deploying** |

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

base_name = "HuggingFaceTB/SmolLM2-135M-Instruct"
tokenizer = AutoTokenizer.from_pretrained(base_name)

# 1) Load base model + the LoRA adapter from notebook 11
base = AutoModelForCausalLM.from_pretrained(base_name)
model = PeftModel.from_pretrained(base, "finetuned_model")

# 2) Merge the LoRA layers INTO the base model -> one normal model
merged = model.merge_and_unload()

# 3) Save the merged model and tokenizer to a folder
merged.save_pretrained("merged_model")
tokenizer.save_pretrained("merged_model")
print("Merged model saved to the 'merged_model' folder.")
```

## Key points to remember

- After training you have **base model + LoRA adapter** (two pieces).
- **`merge_and_unload()`** combines them into **one normal model**.
- Save it with **`save_pretrained("folder")`** (save the **tokenizer** too!).
- A merged model is **easier to share and deploy** - just one folder.
- Keep the adapter separately too if you might want to un-merge later.
