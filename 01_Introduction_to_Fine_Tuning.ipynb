# 01. Introduction to Fine-Tuning

**Fine-tuning** means taking a model that is **already trained** (it already knows language)
and **training it a little more** on *your own examples* so it behaves the way you want.

You are not building a brain from zero. You are **teaching an existing brain your style**.

## Real-life analogy

You hire a **smart new graduate**. They already know English, maths, and general knowledge
(this is the **pre-trained** model).

But they do not know **your company's** way of writing emails. So you give them a folder of
**past examples** and they practise. Now they write exactly like your company.

- Smart graduate = the base model.
- Company examples = your dataset.
- The practice = fine-tuning.

## When should you fine-tune?

| Situation | Fine-tune? |
|-----------|-----------|
| You want a special **style / tone / format** | Yes |
| You have **many good examples** (100s–1000s) | Yes |
| You only need a few facts looked up | No - use RAG |
| You can fix it by writing a better prompt | No - use prompting |
| You have very little data | Usually No |

Fine-tuning is best when you want to **change how the model behaves**, and you have examples to show it.

```python
# A base model already knows language. Let's see it answer BEFORE any fine-tuning.
# (This downloads a small model the first time - it may take a minute.)
from transformers import pipeline

chat = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-135M-Instruct")

messages = [{"role": "user", "content": "Say hello to a new student."}]
out = chat(messages, max_new_tokens=40)
print(out[0]["generated_text"][-1]["content"])
```

## Key points to remember

- **Fine-tuning** = train an **already-trained** model a bit more on **your** examples.
- It changes **how the model behaves** (style, tone, format, task).
- You need **good examples** (a dataset) to teach it.
- Do **not** fine-tune when a better prompt or RAG would solve it.
- The starting model is called the **base model**.
