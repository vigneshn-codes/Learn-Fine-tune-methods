# 03. Understanding the LLM Training Pipeline

A modern chat model is built in **three stages**:

**Pre-training  ->  SFT (Supervised Fine-Tuning)  ->  Alignment**

Knowing these stages tells you **where your fine-tuning fits in** (it is the SFT/Alignment part).

## Real-life analogy (raising a helpful expert)

1. **Pre-training** = **school + reading the whole library**. The model reads huge amounts of text
   and learns language and facts. Very expensive - done once by big labs.
2. **SFT** = **job training**. Show it example *instructions and good answers* so it learns to follow instructions.
3. **Alignment** = **feedback / manners**. Teach it which answers people *prefer* (helpful, safe, polite).

## The three stages

| Stage | What it learns | Data used | Who usually does it |
|-------|----------------|-----------|---------------------|
| **Pre-training** | Language + world knowledge | Huge raw text (internet, books) | Big labs (very costly) |
| **SFT** | To **follow instructions** | Instruction -> answer pairs | **You can do this!** |
| **Alignment** (RLHF / DPO) | To give **preferred** answers | "answer A is better than B" pairs | Advanced teams |

When people say "fine-tuning", they almost always mean **SFT** - and that is what this course focuses on.

```python
# There is no code to "run" a pipeline here - it is a concept.
# This little print just summarises the flow so it is easy to remember.

stages = [
    ("1. Pre-training", "Read huge text  -> learns language & facts"),
    ("2. SFT",          "Instruction/answer pairs -> learns to follow instructions"),
    ("3. Alignment",    "Preference data -> learns which answers people prefer"),
]
for name, what in stages:
    print(f"{name:18} {what}")
```

## Key points to remember

- LLMs are built in **3 stages**: **Pre-training -> SFT -> Alignment**.
- **Pre-training** = learn language from huge text (done by big labs, very costly).
- **SFT** = teach it to follow instructions using example pairs. **This is the fine-tuning you will do.**
- **Alignment** (RLHF/DPO) = teach it which answers people prefer.
- Your dataset of instruction/answer pairs plugs into the **SFT** stage.
