# 12. Monitoring Training & Avoiding Overfitting

While training, we watch the **loss**. We also make sure the model is **learning the pattern**,
not just **memorising** the examples (that is called **overfitting**).

## Real-life analogy

A student who **understands** the subject can answer **new** questions.
A student who only **memorised** the answer sheet fails on new questions. That is **overfitting**.

## Key ideas

| Term | Meaning |
|------|---------|
| **Loss** | How wrong the model is. Lower = better. It should go **down** while training |
| **Training loss** | Error on the data it is **learning from** |
| **Validation loss** | Error on data it has **not seen**. This is the honest score |
| **Overfitting** | Training loss keeps dropping but **validation loss goes UP** |
| **Epoch** | One full pass over the dataset |

### How to avoid overfitting
- Use **fewer epochs** (start with 1-3).
- Keep a **validation set** and watch its loss.
- Use **more / more varied** data.
- Use **LoRA dropout** and a smaller **learning rate**.

```python
# Split data into train and validation so we can watch BOTH losses.
from datasets import Dataset

data = [{"messages": [
            {"role": "user", "content": f"Question {i}?"},
            {"role": "assistant", "content": f"Answer {i}."}]} for i in range(10)]

dataset = Dataset.from_list(data)
split = dataset.train_test_split(test_size=0.2)   # 80% train, 20% validation
print("Train examples:     ", len(split["train"]))
print("Validation examples:", len(split["test"]))

# In SFTConfig you then turn on evaluation to SEE the validation loss:
import torch
from trl import SFTConfig
args = SFTConfig(
    output_dir="out",
    num_train_epochs=3,
    eval_strategy="epoch",     # check validation loss after each epoch
    logging_steps=1,
    report_to="none",
    use_cpu=not torch.cuda.is_available(),   # allow this to build on a CPU laptop too
)
print("\nWatch: if validation loss starts RISING, stop - that is overfitting.")
```

## Key points to remember

- **Loss** = how wrong the model is; it should go **down** while training.
- Keep a **validation set**; the **validation loss** is the honest score.
- **Overfitting** = training loss falls but **validation loss rises** (it memorised).
- Fight overfitting with **fewer epochs**, **more/varied data**, **dropout**, smaller **learning rate**.
- Turn on `eval_strategy="epoch"` to watch validation loss.
