# 02. Fine-Tuning vs Prompt Engineering vs RAG

There are **three ways** to make an LLM do what you want. They are **not** enemies -
you pick the right one for the job (and sometimes mix them).

## The three ways in one line

- **Prompt Engineering** - *ask better*. You write clear instructions and examples in the prompt.
- **RAG** (Retrieval-Augmented Generation) - *give it notes*. You fetch facts and paste them into the prompt.
- **Fine-Tuning** - *train it*. You permanently change the model's behaviour with examples.

## Real-life analogy (a new employee)

- **Prompt Engineering** = giving **clear instructions** for one task.
- **RAG** = handing them the **company handbook** to read while they work.
- **Fine-Tuning** = sending them to **weeks of training** so it becomes second nature.

## Compare them

| | Prompt Engineering | RAG | Fine-Tuning |
|---|---|---|---|
| Changes the model? | No | No | **Yes** |
| Good for | Simple tasks, quick fixes | **Fresh facts**, documents | **Style / behaviour** |
| Needs a dataset? | No | No (needs documents) | **Yes** |
| Cost / effort | Lowest | Medium | Highest |
| Speed to try | Instant | Fast | Slow (training) |

**Rule of thumb:** try **prompting** first -> add **RAG** if it needs facts -> **fine-tune** only if you need a new behaviour.

## Key points to remember

- **Prompt Engineering** = ask better (no training).
- **RAG** = give the model fresh facts/documents to read (no training).
- **Fine-Tuning** = actually train the model to change its **behaviour**.
- Start cheap: **prompt -> RAG -> fine-tune**.
- Fine-tuning is for **style and behaviour**, not for adding fresh facts.
