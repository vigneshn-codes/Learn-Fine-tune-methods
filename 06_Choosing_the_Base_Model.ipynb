# 06. Choosing the Right Base Model

The **base model** is the model you start from. Picking a good one makes fine-tuning
much easier. Let's learn how to choose.

## Real-life analogy

You are hiring someone to train. Do you pick a **fresh graduate** (small, cheap, needs more training)
or an **experienced person** (big, capable, costs more)?

The "right" choice depends on **your task, your budget, and your GPU**.

## What to look at

| Factor | Question to ask |
|--------|-----------------|
| **Size** | How many parameters? (0.5B, 7B, 70B...). Bigger = smarter but needs more GPU |
| **License** | Are you **allowed** to use it commercially? (check the model card) |
| **Base vs Instruct** | **Instruct** models already follow chat; **base** models are raw |
| **Context length** | How much text can it read at once? |
| **Community** | Popular models have more guides and support |

### Popular starting points

| Model | Size | Good for |
|-------|------|----------|
| `HuggingFaceTB/SmolLM2-135M-Instruct` | 135M | Tiny demos, learning (runs on CPU) |
| `TinyLlama/TinyLlama-1.1B-Chat-v1.0` | 1.1B | Light fine-tuning |
| `meta-llama/Llama-3.2-3B-Instruct` | 3B | Strong small model |
| `mistralai/Mistral-7B-Instruct-v0.3` | 7B | Great all-rounder (needs GPU) |

> **Base vs Instruct:** if you want a chatbot, start from an **Instruct** model. If you want full control of the style, a **base** model is fine.

```python
# Load a small base model and see how big it is.
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "HuggingFaceTB/SmolLM2-135M-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Count the parameters (the "size" of the model).
total = sum(p.numel() for p in model.parameters())
print(f"Model: {model_name}")
print(f"Parameters: {total:,}  (~{total/1e6:.0f} million)")
```

## Key points to remember

- The **base model** is your starting point - choose it before anything else.
- Check **size** (fits your GPU?), **license** (allowed to use?), and **base vs instruct**.
- **Instruct** models already chat; **base** models are raw text predictors.
- Start **small** while learning (SmolLM2, TinyLlama), scale up later.
- Bigger models are smarter but need **more GPU memory**.
