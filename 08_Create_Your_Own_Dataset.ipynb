# 08. Creating & Preparing Your Own Dataset

Now let's **make our own** small dataset and save it in the standard **JSONL** format
(one JSON object per line). This is exactly what training tools expect.

## Real-life analogy

You are writing a **flashcard deck**: each card has a **question** and the **perfect answer**.
The model will study these cards. Here we write a few cards and save them to a file.

## Step 1 - Write a few examples (chat format)

```python
# Each example is a short conversation: the user asks, the assistant answers
# in the style we want (here: a friendly support bot for "GreenCart" store).
data = [
    {"messages": [
        {"role": "user", "content": "How long does delivery take?"},
        {"role": "assistant", "content": "Hi! GreenCart delivers within 2-3 business days. "},
    ]},
    {"messages": [
        {"role": "user", "content": "Can I return an item?"},
        {"role": "assistant", "content": "Of course! You can return any item within 7 days for a full refund."},
    ]},
    {"messages": [
        {"role": "user", "content": "Do you deliver on Sundays?"},
        {"role": "assistant", "content": "We currently deliver Monday to Saturday. Sunday delivery is coming soon!"},
    ]},
]
print(f"We have {len(data)} training examples.")
```

## Step 2 - Save to a JSONL file

**JSONL** = one JSON object per line. It is the standard format for training data.

```python
import json

with open("my_dataset.jsonl", "w", encoding="utf-8") as f:
    for row in data:
        f.write(json.dumps(row) + "\n")   # one example per line

print("Saved my_dataset.jsonl")
print("\nFile contents:")
print(open("my_dataset.jsonl", encoding="utf-8").read())
```

## Step 3 - Load it back with the `datasets` library

This is how the training tools will read your data.

```python
# !pip install datasets   (run once if needed)
from datasets import load_dataset

dataset = load_dataset("json", data_files="my_dataset.jsonl", split="train")

print(dataset)
print("\nFirst example:")
print(dataset[0])
```

## Key points to remember

- Save training data as **JSONL** - one JSON example per line.
- Use the **chat** format: a list of `messages` with `role` and `content`.
- Keep the assistant answers in the **exact style** you want the model to learn.
- Load it easily with `load_dataset("json", data_files=..., split="train")`.
- Even **a few clean examples** are enough to start learning the workflow.
