# 09. Tokenization & Chat Templates Explained

Models do **not** read words - they read **tokens** (numbers). **Tokenization** is turning text
into those numbers. A **chat template** arranges a conversation into the exact text the model expects.

## Real-life analogy

- **Tokenization** = translating your sentence into the model's **secret number language**.
- **Chat template** = the **official form** the model wants the conversation written on
  (with special markers for "user said..." and "assistant said...").

## Step 1 - Turn text into tokens (numbers)

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("HuggingFaceTB/SmolLM2-135M-Instruct")

text = "Fine-tuning is fun!"
tokens = tokenizer.encode(text)          # text -> numbers
print("Text:  ", text)
print("Tokens:", tokens)
print("Back to text:", tokenizer.decode(tokens))   # numbers -> text
```

## Step 2 - Apply the chat template

Instead of writing special markers by hand, the tokenizer does it for us with
`apply_chat_template`. This is the **exact text** the model trains and chats on.

```python
messages = [
    {"role": "user", "content": "What is the capital of France?"},
    {"role": "assistant", "content": "The capital of France is Paris."},
]

# tokenize=False -> show it as readable text so we can SEE the format
formatted = tokenizer.apply_chat_template(messages, tokenize=False)
print(formatted)
```

## Why this matters for fine-tuning

During training, every example is turned into tokens using the **chat template**.
If you use the **wrong** template, the model gets confused. Good news: tools like `SFTTrainer`
apply the correct template **for you** - you just provide the `messages`.

## Key points to remember

- Models read **tokens** (numbers), not words - **tokenization** does the conversion.
- `tokenizer.encode(text)` -> numbers; `tokenizer.decode(numbers)` -> text.
- A **chat template** formats a conversation into the exact text the model expects.
- Use `tokenizer.apply_chat_template(messages, ...)` - don't write markers by hand.
- Always use the **tokenizer that matches your model**.
