# 04. Types of Fine-Tuning

There is more than one way to fine-tune. They differ in **how much of the model you change**
and **how much memory (GPU) you need**.

The main ones: **Full Fine-Tuning, PEFT, LoRA, QLoRA**.

## Real-life analogy (updating a house)

- **Full Fine-Tuning** = **rebuild the whole house**. Powerful but slow and expensive.
- **PEFT** (Parameter-Efficient Fine-Tuning) = the **idea** of changing only a *small part*.
- **LoRA** = add **small sticky notes** on top of the house instead of rebuilding. Cheap and fast.
- **QLoRA** = the same sticky notes, but you also **shrink (quantize)** the house to save even more space.

## Compare the types

| Type | What changes | GPU memory | Speed | Quality |
|------|--------------|-----------|-------|---------|
| **Full Fine-Tuning** | **All** weights | Very high | Slow | Highest |
| **PEFT** (family) | A **small** set of weights | Low | Fast | Great |
| **LoRA** | Tiny **added** layers | Low | Fast | Great |
| **QLoRA** | LoRA **+ 4-bit model** | **Lowest** | Fast | Great |

- **PEFT** is the family name. **LoRA** and **QLoRA** are the most popular members.
- **QLoRA** lets you fine-tune big models on a **single small GPU** - that is why it is so loved.

```python
# A LoRA setup is just a small config object. (We only DESCRIBE it here;
# we actually use it in later notebooks.)
# LoRA adds a few tiny trainable layers instead of changing the whole model.

lora_settings = {
    "r": 8,                 # size of the "sticky note" layers (rank)
    "lora_alpha": 16,       # how strongly the sticky notes are applied
    "target_modules": ["q_proj", "v_proj"],  # where to attach them
    "lora_dropout": 0.05,   # a little randomness to avoid overfitting
}
for k, v in lora_settings.items():
    print(f"{k:16} = {v}")
```

## Key points to remember

- **Full fine-tuning** changes **all** weights - best quality but needs a **big GPU**.
- **PEFT** = the family of methods that change only a **small** part of the model.
- **LoRA** adds **tiny extra layers** and trains only those - cheap and fast.
- **QLoRA** = LoRA **+ a 4-bit (shrunk) model** = the **lowest memory** option.
- For most people: **LoRA or QLoRA** is the right choice.
