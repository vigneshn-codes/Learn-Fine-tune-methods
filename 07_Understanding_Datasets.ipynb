# 07. Understanding Datasets for Fine-Tuning

The model learns from **examples**. A good dataset is the **most important** part of fine-tuning.
"Garbage in, garbage out" - great data makes a great model.

## Real-life analogy

To teach someone your writing style, you give them **many examples** of
"here is a question, here is the perfect answer". The model learns the **pattern** from these pairs.

## Common dataset formats

| Format | Looks like | When to use |
|--------|-----------|-------------|
| **Instruction** | `{"instruction": ..., "output": ...}` | Task-style tuning |
| **Chat** | `{"messages": [{"role": "user", ...}, {"role": "assistant", ...}]}` | Chatbots (most common today) |
| **Completion** | `{"prompt": ..., "completion": ...}` | Simple text continuation |

The **chat** format is the most popular because most models are chat models.

### What makes a good dataset
- **Quality over quantity** - a few hundred *clean* examples beat thousands of messy ones.
- **Consistent style** - answers should all follow the tone you want.
- **Variety** - cover the different kinds of questions users will ask.

```python
# Let's look at ONE example in the popular "chat" format.
example = {
    "messages": [
        {"role": "user", "content": "What is the capital of France?"},
        {"role": "assistant", "content": "The capital of France is Paris."},
    ]
}

exmp2 ={"instruction": "What is the capital of France?", "output": "The capital of France is Paris."}


for m in example["messages"]:
    
    print(f"{m['role']:10}: {m['content']}")
```

## Key points to remember

- The dataset is **the most important part** - quality data = quality model.
- Common formats: **instruction**, **chat** (messages), and **completion**.
- The **chat** format (`messages` with roles) is the most used today.
- Aim for **clean, consistent, varied** examples.
- **Quality beats quantity** - clean data matters more than huge data.
