# 13. Evaluating the Fine-Tuned Model

After training, we must check: **did it actually get better?**
We evaluate in two ways: **look at the answers** (qualitative) and **measure a score** (quantitative).

## Real-life analogy

After training an employee, you give them a **test**. You read their answers (do they sound right?)
and you also give a **score** (how many correct?). Same for our model.

## Two ways to evaluate

| Way | What you do | Good for |
|-----|-------------|----------|
| **Qualitative** | Read the model's answers before vs after | Style, tone, "does it sound right?" |
| **Quantitative** | Measure a number (e.g. **validation loss**, **perplexity**) | Comparing versions objectively |

**Perplexity** = a number for "how surprised the model is by the text". **Lower is better.**

## Qualitative check: compare answers

Load your fine-tuned adapter and ask the same question you trained on. Does it answer in the new style?

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from peft import PeftModel

base_name = "HuggingFaceTB/SmolLM2-135M-Instruct"
tokenizer = AutoTokenizer.from_pretrained(base_name)

# Load base model, then attach the LoRA adapter we trained in notebook 11.
base = AutoModelForCausalLM.from_pretrained(base_name)
model = PeftModel.from_pretrained(base, "finetuned_model")

# Wrap it in a pipeline so we can just pass the chat messages and read the answer.
chat = pipeline("text-generation", model=model, tokenizer=tokenizer)

messages = [{"role": "user", "content": "How long does delivery take?"}]
out = chat(messages, max_new_tokens=40)
print(out[0]["generated_text"][-1]["content"])
```

## Quantitative check: the idea of perplexity

You do not need to memorise the maths. Just remember: **run the model on a test set and read the score.**
A lower **validation loss** or **perplexity** after fine-tuning means it improved.

## Key points to remember

- Always check: **did fine-tuning actually help?**
- **Qualitative** = read answers before vs after (style, correctness).
- **Quantitative** = a number like **validation loss** or **perplexity** (**lower is better**).
- Test on questions the model **did not train on** for an honest result.
- If it did not improve, fix the **data** or the **settings** and try again.
