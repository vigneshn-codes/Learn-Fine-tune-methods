# 15. Running Inference with the Fine-Tuned Model

**Inference** = **using** the trained model to get answers (no more training).
Let's load our merged model and chat with it.

## Real-life analogy

Training was **school**. Inference is the employee **doing the actual job** - answering real questions.

## Load the merged model and ask a question

```python
from transformers import pipeline

# Load the merged model we saved in notebook 14.
chat = pipeline("text-generation", model="merged_model")

messages = [{"role": "user", "content": "Can I return an item?"}]
out = chat(messages, max_new_tokens=50)
print(out[0]["generated_text"][-1]["content"])
```

## A few useful generation settings

| Setting | What it does |
|---------|--------------|
| `max_new_tokens` | How long the answer can be |
| `temperature` | Higher = more creative, lower = more focused |
| `do_sample` | `True` = varied answers, `False` = most likely answer |

```python
# Same model, but control the style of the answer.
out = chat(
    [{"role": "user", "content": "Do you deliver on Sundays?"}],
    max_new_tokens=50,
    temperature=0.3,     # low = focused and consistent
    do_sample=True,
)
print(out[0]["generated_text"][-1]["content"])
```

## Key points to remember

- **Inference** = using the trained model to answer (no training).
- The easiest way is the **`pipeline("text-generation", model=...)`** helper.
- Control answers with **`max_new_tokens`**, **`temperature`**, and **`do_sample`**.
- Point it at your **merged_model** folder from the previous notebook.
- Lower **temperature** = more focused; higher = more creative.
